```python
import numpy as np
import pywt
import matplotlib.pyplot as plt

# Create a synthetic signal
x = np.sin(2 * np.pi * 7 * np.linspace(0, 1, 16)) + np.sin(2 * np.pi * 13 * np.linspace(0, 1, 16))
n = len(x)

# Perform a Discrete Wavelet Transform using Haar wavelets
coeffs = pywt.dwt(x, 'haar')
cA, cD = coeffs

# Inverse Discrete Wavelet Transform
reconstructed_signal = pywt.idwt(cA, cD, 'haar')

# Visualization
plt.figure(figsize=(12, 8))

# Original Signal
plt.subplot(4, 1, 1)
plt.title("Original Signal")
plt.plot(x)

# Approximation Coefficients
plt.subplot(4, 1, 2)
plt.title("Approximation Coefficients")
plt.plot(cA)

# Detail Coefficients
plt.subplot(4, 1, 3)
plt.title("Detail Coefficients")
plt.plot(cD)

# Reconstructed Signal
plt.subplot(4, 1, 4)
plt.title("Reconstructed Signal")
plt.plot(reconstructed_signal)

plt.tight_layout()
plt.show()
```
